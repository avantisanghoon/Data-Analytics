```python
import sqlite3
import pandas as pd
```

```python
# 데이터 접속하기 
# 책을 빌리기 위해서는 먼저 도서관에 가야한다.

connection = sqlite3.connect('SQLite/chinook.db') # DB에 접근하는 키
```

* select : 데이터를 찾을때! 데이터 테이블을 불러올때!
* insert : 데이터를 집어넣을때!
* update : 데이터를 갱신할때!
* delete : 데이터를 삭제할때! (많이 사용하지 않음 / 회사에서는 데이터를 쉽게 삭제하지 않음)

```python
# index 사용하기

query = "SELECT * FROM 'invoices'"
df_invoices = pd.read_sql_query(query, connection)
df_invoices.shape
```

```python
df_invoices
```

```python
# python 에서 조건을 걸어서 데이터 가져오기
cond1 = (df_invoices['BillingCountry']=='Germany')
df_invoices.loc[cond1]
```

```python
# query 자체에 조건을 건다.
# query 문 자체에서 미리 조건을 건다.
query = "SELECT * FROM 'invoices' WHERE BillingCountry='Germany'"
df_invoices = pd.read_sql_query(query, connection)
df_invoices.shape
```
